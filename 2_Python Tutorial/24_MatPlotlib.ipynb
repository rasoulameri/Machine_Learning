#### Matplotlib Pyplot

````python
# Draw a line in a diagram from position (1,1) to position (10,100):
import matplotlib.pyplot as plt # Now the Pyplot package can be referred to as plt
import numpy as np

xpoints = np.array([1, 10])
ypoints = np.array([1, 100])

plt.figure(figsize=(9,6))
plt.plot(xpoints, ypoints)
plt.show()
````

````python
# Draw two points in the diagram,  one at position (1, 1) and one in position (10, 100)
xpoints = np.array([1, 10])
ypoints = np.array([1, 100])

plt.figure(figsize=(9,6))
plt.plot(xpoints, ypoints, 'o')
plt.show()
````

````python
xpoints = np.array([1, 2, 4, 5])
ypoints = np.array([3, 4, 1, 4])

plt.figure(figsize=(9,6))
plt.plot(xpoints, ypoints)
plt.show()
````

````python
import numpy as np
x = np.arange(0, 10, 0.2)
y = np.sin(x)

plt.plot(x, y)
plt.show()
````

````python
ypoints = np.array([2, 8, 3, 10])

plt.figure(figsize=(9,6))
plt.plot(ypoints, marker = 'o') #marker = o
plt.show()
````

````python
# add title and label
plt.figure(figsize=(9,6))
plt.plot(ypoints)

plt.title("Title")
plt.xlabel("Axis_X")
plt.ylabel("Axis_Y")

plt.show()
````

````python
# define the fonts for title and axis labels
plt.figure(figsize=(9,6))
plt.plot(ypoints)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Title", fontdict= font1)
plt.xlabel("Axis_X", fontdict= font2)
plt.ylabel("Axis_Y", fontdict= font2)

plt.show()
````

````python
# Grid on
plt.figure(figsize=(9,6))
plt.plot(ypoints)

plt.title("Title")
plt.xlabel("Axis_X")
plt.ylabel("Axis_Y")

plt.grid() # grid on
#plt.grid(axis = 'x')
#plt.grid(axis = 'y')


plt.show()
````

#### Subplot

````python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([2, 4, 6, 8])

plt.figure(figsize=(9,6))
plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([3, 6, 9, 12])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
````

````python
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5, label='Normal')
ax.plot(x15, x15*3.0, label='Fast')
ax.plot(x15, x15/3.0, label='Slow')

ax.legend();
````

### Scatter Plots

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
````

````python
# Scotter plot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.figure(figsize=(9,6))

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
````

````python
# Bars
x = np.array(["A", "B", "C", "D"])
y = np.array([2, 8, 1, 12])

plt.figure(figsize=(9,6))
plt.bar(x,y)
plt.show()
````

````python
# Histogram
x = np.random.normal(180, 10, 200) #A Normal Data Distribution by NumPy

print(x)
````

````python
x = np.random.normal(180, 10, 200)

plt.figure(figsize=(9,6))
plt.hist(x)
plt.show() 
````

````python
# Pie Charts
y = np.array([30, 40, 10, 20])
mylabels = ["A", "B", "C", "D"]

plt.figure(figsize=(9,6))
plt.pie(y, labels=mylabels)
plt.show() 
````

````python
# Explode and legend
y = np.array([30, 40, 10, 20])
mylabels = ["A", "B", "C", "D"]

myexplode = [0.2, 0, 0, 0]

plt.figure(figsize=(9,6))
plt.pie(y, labels = mylabels, explode = myexplode)
plt.legend()
plt.savefig('test.png', dpi = 1024) # save the results
plt.show() 
````
