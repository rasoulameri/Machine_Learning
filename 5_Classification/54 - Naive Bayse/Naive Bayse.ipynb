#### <span style='color:Black'> **Introduction** </span>

- It is a statistical classification technique based on Bayes Theorem.
- It is one of the simplest supervised learning algorithms.
- Naive Bayes classifier is the fast, accurate algorithm, and has high accuracy on large dataset.
- It is successfully used in various application such as spam filtering, text classification, and recommender systems.

**What is Naive Bayes Classifier?**
- Naive Bayes classifier assumes that the effect of a particular feature in a class is independent of other features.

**Bayse theorem:**

<span style='color:Blue'>$$P(h|D) = \frac{P(D|h)P(h)}{P(D)} $$</span>
<br>

- P(h): the probability of hypothesis h being true (regardless of the data). This is known as the prior probability of h.
- P(D): the probability of the data (regardless of the hypothesis). This is known as the prior probability.
- P(h|D): the probability of hypothesis h given the data D. This is known as posterior probability.
- P(D|h): the probability of data d given that the hypothesis h was true. This is known as posterior probability.

<br> 

#### <span style='color:Black'> **How does Naive Bayse work?** </span>

<span style='color:darkgreen'> **Example:** Given an example of weather conditions and playing sports. You need to calculate the probability of playing sports.</span>

<span style='color:darkgreen'>**Step 1:** Convert the data set into a frequency table</span>

<span style='color:darkgreen'>**Step 2:** Create Likelihood table by finding the probabilities</span>

    <img src='./imgs/pic1.png'>
</p>

<span style='color:darkgreen'> **Step 3:** Now, use Naive Bayesian equation to calculate the posterior probability for each class. The class with the highest posterior probability is the outcome of prediction. </span>

**Problem: Players will play if weather is sunny. Is this statement is correct?**
$$P(Yes | Sunny) = \frac{P( Sunny | Yes) \times P(Yes)} {P (Sunny)} $$

<span style='color:darkgreen'> We have:<br> $P (Sunny |Yes) = \frac{3}{9} = 0.33$,<br> $P(Sunny) = \frac{5}{14} = 0.36$,<br> $P( Yes)= \frac{9}{14} = 0.64$ </span>
<br><br>
<span style='color:darkgreen'> Now,  $P (Yes | Sunny) = \frac{0.33 * 0.64}{0.36} = 0.60$, which has higher probability. </span>

#### <span style='color:Black'>**Naive Bayse classifier in Python**</span>

<br> <span style = 'color:RebeccaPurple'>**Import libraries**</span>

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
```

<br> <span style = 'color:RebeccaPurple'>**Importing the dataset**</span>

```python
dataset = pd.read_csv('./Data/Social_Network_Ads.csv')
dataset.head()
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
print(len(X))
```

<br> <span style = 'color:RebeccaPurple'>**Splitting the dataset into the Training set and Test set**</span>

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
print(len(X_train))
print(len(X_test))
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

<br> <span style = 'color:RebeccaPurple'>**Feature Scaling**</span>

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

<br> <span style = 'color:RebeccaPurple'>**Training the Naive Bayes model on the Training set**</span>

```python
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

<br> <span style = 'color:RebeccaPurple'>**Predicting a new result**</span>

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

<br> <span style = 'color:RebeccaPurple'>**Predicting the Test set results**</span>

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

<br> <span style = 'color:RebeccaPurple'>**Making the Confusion Matrix**</span>

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

<br> <span style = 'color:RebeccaPurple'>**Visualising the Training set results**</span>

```python
X_set, y_set = sc.inverse_transform(X_train), y_train

X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))

plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
    
plt.title('Naive Bayes (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

<br> <span style = 'color:RebeccaPurple'>**Visualising the Test set results**</span>

```python
X_set, y_set = sc.inverse_transform(X_test), y_test

X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))

plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
    
plt.title('Naive Bayes (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

#### <span style='color:Black'> **Zero Probability Problem** </span>


Suppose there is no tuple for a condition in the dataset, in this scenario, the posterior probability will be zero, and the model is unable to make a prediction. 

**This problem is known as Zero Probability because the occurrence of the particular class is zero.**

**SOLUTION**
The solution for such an issue is the Laplacian correction or Laplace Transformation. Laplacian correction is one of the smoothing techniques. Here, you can assume that the dataset is large enough that adding one row of each class will not make a difference in the estimated probability. This will overcome the issue of probability values to zero.
